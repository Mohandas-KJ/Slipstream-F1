# Telemetry Analysis

```python
#Imports
import sys
import importlib
sys.path.append("../..")
import Sliplib.slipload as sl
import Sliplib.selection as slipsel
import Sliplib.plotting as slpolt
import Sliplib.dashboard as db

importlib.reload(slpolt)
```

```python
# Select GP and Load Session
session = sl.load_gp(2025,"Spanish Grand Prix","Race")
session.load()
```

```python
# Extract Lap Data
laps = session.laps
laps.head()
```

```python
# Load Drivers (Max and Lando)
ver,nor = slipsel.pick_athletes(laps,["VER","NOR"])
```

```python
# Verstappen Data
ver.head(5)
```

```python
# Norris Data
nor.head(5)
```

```python
# Pick the Fastest Laps
ver_fast = ver.pick_fastest()
nor_fast = nor.pick_fastest()
```

```python
# Verify
ver_fast
```

```python
# Verstappen's Fastest Lap time
ver_fast.LapTime
```

```python
# Norris's Fastest Lap time
nor_fast.LapTime
```

```python
# Get Car data (telemetry Data)
ver_tel = ver_fast.get_car_data().add_distance()
nor_tel = nor_fast.get_car_data().add_distance()
```

```python
# Verify
nor_tel
```

```python
# Know data
ver_tel.columns
```

```python
# Plot of Distance vs Speed
slpolt.plot_dist_speed([ver_tel,nor_tel],["VER","NOR"])
```

```python
# Extract Data to .csv for analysis
ver_tel.to_csv("../data/VER_Telemetry.csv")
```

```python
# Plot Distance vs Brake
slpolt.plot_dist_brake([ver_tel,nor_tel],["VER","NOR"])
```

```python
# Track Layout
import matplotlib.pyplot as plt

tel = ver_fast.get_telemetry()

plt.plot(
    tel["X"],
    tel["Y"]
)
```

```python
# Plot Throttle vs Distance
slpolt.plot_dist_throttle([ver_tel,nor_tel],["VER","NOR"])
```

```python
# Plotting Distance vs nGear to analyse the timing of shifts
slpolt.plot_dist_ngear([ver_tel,nor_tel],["VER","NOR"])
```

```python
# Plotting Distance vs RPM of Drivers
slpolt.plot_dist_rpms([ver_tel,nor_tel],["VER","NOR"])
```
