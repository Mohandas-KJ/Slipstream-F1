# Telemetry Analysis

```python
#Imports
import sys
import importlib
sys.path.append("../..")
import Sliplib.slipload as sl
import Sliplib.selection as slipsel
import Sliplib.plotting as slpolt

importlib.reload(slpolt)
```

```python
# Select GP and Load Session
session = sl.load_gp(2025,"Spanish Grand Prix","Race")
session.load()
```

```python
# Extract Lap Data
laps = session.laps
laps.head()
```

```python
# Load Drivers (Max and Lando)
ver,nor = slipsel.pick_athletes(laps,["VER","NOR"])
```

```python
# Verstappen Data
ver.head(5)
```

```python
# Norris Data
nor.head(5)
```

```python
# Pick the Fastest Laps
ver_fast = ver.pick_fastest()
nor_fast = nor.pick_fastest()
```

```python
# Verify
ver_fast
```

```python
# Get Car data (telemetry Data)
ver_tel = ver_fast.get_car_data().add_distance()
nor_tel = nor_fast.get_car_data().add_distance()
```

```python
# Verify
nor_tel
```

```python
# Know data
ver_tel.columns
```

```python
# Plot of Distance vs Speed
slpolt.plot_dist_speed([ver_tel,nor_tel],["VER","NOR"])
```
